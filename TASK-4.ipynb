## Data Cleaning

```python
# extract only required columns from the original dataset.
df=df[['iyear','country_txt','region_txt','city','attacktype1_txt','targtype1_txt','gname','nkill']]
df.head()
```

# Data Visualization

```python
# year wise terrorist attack.
import matplotlib.pyplot as plt
import seaborn as sns
plt.subplots(figsize=(25,8))
sns.countplot(x='year', data=df,ec='black')
plt.title("Number of terrorist attacks year wise ")
plt.xticks(rotation=90)
plt.xlabel("Year",fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.show()
```

```python
# region wise and year wise terrorist attcks visualization.
pd.crosstab(df.year,df.region).plot(kind='line',figsize=(20,6))
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Terrorist Activities by Region in each Year')
plt.show()
```

```python
# Group the data by country and count the number of attacks
attacks_by_country = df.groupby('country')['country'].count().sort_values(ascending=False)[:10]

# Plot the data
plt.figure(figsize=(12, 6))
sns.barplot(x=attacks_by_country.index, y=attacks_by_country)
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.title('Top 10 Countries with the Highest Number of Terrorist Attacks')
plt.show()
```

```python
# Group the data by attack type and count the number of attacks
attacks_by_type = df.groupby('attacktype')['attacktype'].count()

# Plot the data
plt.figure(figsize=(25,10))
sns.barplot(x=attacks_by_type.index, y=attacks_by_type)
plt.xlabel('Type of Attack')
plt.ylabel('Number of Attacks')
plt.title('Types of Attack Used by Terrorists')
plt.show()
```

```python
# Group the data by target and count the number of attacks
attacks_by_target = df.groupby('target')['target'].count().sort_values(ascending=False)[:10]

# Plot the data
plt.figure(figsize=(25,10))
sns.barplot(x=attacks_by_target.index, y=attacks_by_target)
plt.xlabel('Target')
plt.ylabel('Number of Attacks')
plt.title('Top 10 Targets of Terrorist Attacks')
plt.show()
```

```python
df['attacktype'].value_counts().plot(kind='pie', figsize=(5,10), autopct='%1.1f%%')
plt.title('Proportion of Terrorist Attacks by Type')
plt.show()
```

```python
df.groupby('country')['killed'].sum().sort_values(ascending=False).head(20).plot(kind='bar', figsize=(15,5))
plt.xlabel('Country')
plt.ylabel('Number of People Killed')
plt.title('Top 20 Countries with the Highest Number of People Killed in Terrorist Attacks in each country')
plt.show()
```

```python
df.groupby('organization')['killed'].sum().sort_values(ascending=False).head(20).plot(kind='bar', figsize=(15,5))
plt.xlabel('Organization')
plt.ylabel('Number of People Killed')
plt.title('Top 20 Terrorist Organizations with the Highest Number of People Killed in each orginisation')
plt.show()
```

```python
target_counts = df['target'].value_counts().head(10)

plt.figure(figsize=(12,6))
plt.pie(target_counts.values, labels=target_counts.index)
plt.title('Terrorist attacks by target type (top 10)')
plt.show()
```

```python
df['count'] = 1
year_attacktype_counts = df.groupby(['year', 'attacktype'])['count'].sum()
year_attacktype_counts = year_attacktype_counts.unstack()
plt.figure(figsize=(25,8))
sns.pointplot(data=year_attacktype_counts)
plt.xlabel('Year')
plt.ylabel('Number of terrorist attacks')
plt.title('Terrorist attacks by attack type and year')
plt.show()
```
